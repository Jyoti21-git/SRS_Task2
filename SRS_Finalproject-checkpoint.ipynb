# Content Based Filtering Model

```python
df_sampled = genre_data.sample(n=10000, random_state=42).reset_index(drop=True)  
```

```python
print(df_sampled)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors as NN
```

# To select those columns which are desired to find the relation between the songs.

```python
features = ['danceability', 'key', 'mode', 'loudness', 'tempo', 'valence', 'energy']
X = df_sampled[features]

```

# Normalising/standardising the features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

# Create a NearestNeighbors model


```python
m = NN(n_neighbors=6, algorithm='auto', metric='euclidean')  
model.fit(X_scaled)

# Function to get recommendations using NearestNeighbors
def get_recommendations(song_name):
    # Check if the song exists in the sampled DataFrame
    if song_name not in df_sampled['song_name'].values:
        print(f"Song '{song_name}' not found in the dataset.")
        return None
    
    idx = df_sampled[df_sampled['song_name'] == song_name].index[0]
    distances, indices = model.kneighbors(X_scaled[idx:idx + 1])  # Get neighbors
    return df_sampled.iloc[indices[0][1:]]  
```

```python
# Testing
if __name__ == "__main__":
    song_to_recommend = input("Enter the song: ") 
    recommendations = get_recommendations(song_to_recommend)
    if recommendations is not None:
        song_names = recommendations['song_name'] if isinstance(recommendations, p.DataFrame) else recommendations
        print(song_names.tolist()) 
```
